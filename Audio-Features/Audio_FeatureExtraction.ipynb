### Extract Features from the generated .wav using SMILExtract frame wise

```python
import subprocess

dir = "C:/Users/Tilak Chan/OneDrive/Desktop/Tilak/BagOfLies/wavfiles"
#change outdir="/home/adrikamukherjee/Data_available/Audio_features/arff_files_frame_wise/" for other datasets
outdir="C:/Users/Tilak Chan/OneDrive/Desktop/Tilak/BagOfLies/Audio_features/arff_files_frame_wise/"
list_wav=[]
wav_file_count=0
t = 1
for filename in glob.glob(os.path.join(dir, '*.wav')):
    list_wav.append(filename)
    cmd=""
    input_filename=os.path.basename(filename)
    output_filename=input_filename.replace(".wav",".arff")
    cmd="./bin/SMILExtract -C C:/openSMILE/config/emobase/emobase.conf -I " + "\"" + "C:/Users/Tilak Chan/OneDrive/Desktop/Tilak/BagOfLies/wavfiles/" + input_filename + "\"" + " -O " + "\"" + "C:/Users/Tilak Chan/OneDrive/Desktop/Tilak/BagOfLies/Audio_features/arff_files_frame_wise/" + output_filename + "\""
    cmd="cd C:\openSMILE; "+cmd
    x = subprocess.run(cmd, shell=True)
    print()
    print(cmd)
    t+=1
    if(x==0):
      wav_file_count+=1  
print(wav_file_count)
```
